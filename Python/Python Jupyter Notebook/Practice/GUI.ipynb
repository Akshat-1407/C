```python
# entry widget = textbox that accepts a single line of user input

from tkinter import *

def submit():
    user_input = entry_box.get()
    print("Hello", user_input)
    entry_box.config(state=DISABLED)
    
def delete():
    entry_box.delete(0, END)

def backspace():
    entry_box.delete(len(entry_box.get())-1, END)

window = Tk()
window.title("Entry Box")

entry_box = Entry(window, font=("Arial", 30), fg="green", bg="black", show="*")
# entry_box.insert(0,'Spongebob')  #-  used to insert defaul text
# entry_box.config(show="*")  -  used to show any symbol in place of the letters we are typing
# we can use both these function in the constructor (Entry())

entry_box.pack(side=LEFT)

submit_button = Button(window, command=submit, text="SUBMIT", font=("Arial", 10))
submit_button.pack(side=RIGHT)

delete_button = Button(window, command=delete, text="DELETE", font=("Arial", 10))
delete_button.pack(side=RIGHT)

backspace_button = Button(window, command=backspace, text="BACKSPACE", font=("Arial", 10))
backspace_button.pack(side=RIGHT)

window.mainloop()
```

```python
# radio button = similar to checkbox, but you can only select one from a group

from tkinter import *

food = ["Pizza","Hamburger","Hotdog"]

def order():
    if(x.get()==0):
        print("You ordered Pizza!")
    elif(x.get()==1):
        print("You ordered a Hamburger!")
    elif(x.get()==2):
        print("You ordered a Hotdog!")
    else:
        print("huh?")

window = Tk()
window.title("Radio Button")

pizzaImage = PhotoImage(file='F:\\GUI Images\pizza.png')
hamburgerImage = PhotoImage(file='F:\\GUI Images\hamburger.png')
hotdogImage = PhotoImage(file='F:\\GUI Images\hotdog.png')
foodImages = [pizzaImage,hamburgerImage,hotdogImage]

x = IntVar()

for index in range(len(food)):
    radiobutton = Radiobutton(window,
                              text=food[index],           # adds text to radio buttons
                              variable=x,                 # groups radiobuttons together if they share the same variable
                              value=index,                # assigns each radiobutton a different value
                              padx = 25,                  # adds padding on x-axis
                              font=("Impact",50),
                              image = foodImages[index],  # adds image to radiobutton
                              compound = 'left',          # adds image & text (left-side)
                              #indicatoron=0,             # eliminate circle indicators
                              #width = 375,               # sets width of radio buttons
                              command=order)              # set command of radiobutton to function
    radiobutton.pack(anchor=W) #(west)
window.mainloop()
```

```python
# Colour Chooser

from tkinter import *
from tkinter import colorchooser   # it's a submodule in tkinter

window = Tk()
window.geometry("500x500")

def changeBg():
    colour = colorchooser.askcolor()  # it gives a tuple in which first element is the rgb values and the second element is the hex value of the chosen colour.
    print(colour)
    colourHex = colour[1]
    window.config(background=colourHex)

change_bg = Button(window, text="CLICK TO CHANGE THE BACKGROUND COLOUR",command=changeBg, font=("Constantia", 15))
change_bg.pack()

window.mainloop()
```

```python
# Open file dialog box.

from tkinter import *
from tkinter import filedialog

def openFile():
    # filedialog.askopenfilename  returns a string which is the path of the directiory
    filepath = filedialog.askopenfilename(initialdir="F:\\VS Code",
                                          title="Open File?",
                                          filetypes= (("text files","*.txt"),
                                                      ("all files","*.*")))
    file = open(filepath,'r')  # r is used for read, we can also use rt(read text)
    print(file.read())
    file.close()

window = Tk()
window.title = "GUI Open File"
button = Button(text="OPEN", font=("Arial", 10, 'bold'), command=openFile)
button.pack()
window.mainloop()
```

```python
# write and save a file.

from tkinter import *
from tkinter import filedialog

def saveFile():
    file = filedialog.asksaveasfile(initialdir="F:\\VS Code",
                                    defaultextension='.txt',
                                    filetypes=(("Text file",".txt"),
                                               ("HTML file", ".html"),
                                               ("All files", ".*")))
    if file is None:   # used to prevent an error if we do not save a file.
        return
    fileText = str(text.get(1.0,END))
    # filetext = input("Enter some text: ") //use this if you want to use console window
    file.write(fileText)
    file.close()

window = Tk()
window.geometry("500x500")

button = Button(window, text="SAVE", font=("Arial", 15, 'bold'), command=saveFile)
button.pack(side=BOTTOM)

text = Text(window, bg="light yellow", fg="Purple", font=("Arial", 15, 'italic'), padx=10, pady=10)
text.pack()

window.mainloop()
```

```python
# how to add window tabs?

from tkinter import *
from tkinter import ttk

window = Tk()
window.title("Window Tabs")

notebook = ttk.Notebook(window)  #widget that manages a collection of windows/displays

tab1 = Frame(notebook)  #new frame for tab 1
tab2 = Frame(notebook)  #new frame for tab 2

Text(tab1, font=("Arial", 15), bg="light yellow", fg="black").pack()
Text(tab2, font=("Arial", 15), bg="light yellow", fg="black").pack()

notebook.add(tab1,text="Tab 1")
notebook.add(tab2,text="Tab 2")
notebook.pack(expand=True,fill="both")  #expand = expand to fill any space not otherwise used
                                        #fill = fill space on x and y axis. we can use x, y, both in fill

window.mainloop()
```

```python
# Progress Bar

from tkinter import *
from tkinter.ttk import *
import time

def start():
    GB = 100
    download = 0
    speed = 1
    while(download<GB):
        time.sleep(0.05)
        bar['value']+=(speed/GB)*100
        download+=speed
        percent.set(str(int((download/GB)*100))+"%")
        text.set(str(download)+"/"+str(GB)+" GB completed")
        window.update_idletasks()
    window.destroy()

window = Tk()

percent = StringVar()
text = StringVar()

bar = Progressbar(window,orient=HORIZONTAL,length=300)
bar.pack(pady=10)

percentLabel = Label(window,textvariable=percent).pack()
taskLabel = Label(window,textvariable=text).pack()

button = Button(window,text="download",command=start).pack()

window.mainloop()
```

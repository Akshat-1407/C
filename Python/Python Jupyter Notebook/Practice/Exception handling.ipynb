```python
# Exception Handiling: we use this to handle exceptions and prevent python to raise an error.

try:
    numerator = int(input("Enter an number to divide: "))
    denominator = int(input("Enter an number to divide by: "))
    result = numerator/denominator
except ZeroDivisionError as e:
    print(e)
    print("You can't divide by zero.")
except ValueError as e:
    print(e)
    print("Enter number only.")
except Exception as e:
    print(e)
    print("Something went wrong.")
else:
    print(result)
finally:
    print("This statement will be printed irrespective of whether there is an exception or not")
```

```python
# Raising custom errors = we can raise error by using raise keyword even if there is no error.

a = int(input("Enter a number between 1-10 : "))
if a<1 or a>10:
    raise ValueError("Number should be between 1-10")
```

```python
# Indexing : In Python Indexing starts from 0. For example if we take a word "Psychology" P-0, S-1, Y-2, C-3, .... Y-9.
# Slicing : When we define range during slicing; [x:y] where x is the starting point and y is the ending point whose value is (n-1)

Word="Psychology"
print(Word[8])
print(Word[0:5])
print(Word[3:8])
print(Word[:8]) # If Starting point is not defined it take 0 as the starting point by default.
print(Word[3:],end="\n\n") # If End point is not defined it takes the last indexing as the ending point by default.

print(Word[0:-3])
print(Word[0:len(Word)-3]) # len(Word)=10
print(Word[0:7],end="\n\n")

print(Word[-5:-3])
print(Word[len(Word)-5:len(Word)-3])
print(Word[5:7])
```

```python
# String methods
# Python provides a set of built-in methods that we can use to alter and modify the strings. String are immutable.

# .upper() : It is use to make a new string in which all letter are in uppercase.
# .lower() : It is used to make a new string in which all letter are in lowercase.
# .strip() : It is used to remove any spaces befor or after the string.
# .rstrip() : It is ued to remove any traling character after the string.
# .replace() : It is used to replace any word in a string with an another word.
# .split() : It splits the given string at a specified instance and returns the separated strings as list items.
# .capitalize : It is use to conver the first letter of the string as a capital letter and other as small letters.
# .center() : It is use to align the string to the center.
# .count() : No of times the given value has occured in the same string.
# .endswith() : This method checks if the string ends with a given value. If yes then return True, else return False. We can even also check for a value in-between the string by providing start and end index positions.
# .find() : This method searches for the first occurrence of the given value and returns the index where it is present. If given value is absent from the string then return -1.
# .index() : This method searches for the first occurrence of the given value and returns the index where it is present. If given value is absent from the string then raise an error.
# .isalnum() : This method returns True only if the entire string only consists of A-Z, a-z, 0-9. If any other characters or punctuations are present, then it returns False.
# .isalpha() : This method returns True only if the entire string only consists of A-Z, a-z. If any other characters or punctuations or numbers(0-9) are present, then it returns False.
# .islower() : This method returns True if all the characters in the string are lower case, else it returns False.
# .isprintable() : This method returns True if all the values within the given string are printable, if not, then return False.
# .isspace() : This method returns True only and only if the string contains only white spaces, else returns False.
# .istitile() : This returns True only if the first letter of each word of the string is capitalized, else it returns False.
# .isupper() : This method returns True if all the characters in the string are upper case, else it returns False.
# .startswith() : This method checks if the string starts with a given value. If yes then return True, else return False.
# .swapcase() : This method changes the character casing of the string. Upper case are converted to lower case and lower case to upper case.
# .title() : This method capitalizes each letter of the word within the string.

letter="Windmill"
print(letter.upper())
print(letter.lower())
print("  hello   ".strip())
print("hello!!!!!!".rstrip("!"))
print("How are you.".replace("How","What"))
print("How are you".split(" "))
print("IntRoducton To pytHon.".capitalize(),end="\n\n")

print(letter.center(50))
print(letter.center(50).strip())
print(letter.count("i"),end="\n\n")

print(letter.endswith("l"))
print(letter.endswith("i"))
print("Welcome to the Console".endswith("to",4,10), end="\n\n")

print("He's name is Dan. He is an honest man.".find("is"))
print("He's name is Dan. He is an honest man.".find("Danial"),end="\n\n")

print("He's name is Dan. He is an honest man.".index("is"),end="\n\n")

print("hi20".isalnum())
print("hi 20".isalnum(),end="\n\n")

print("hello".isalpha())
print("how are you".islower())
print("How are you\n".isprintable(),end='\n\n')

print(" hello ".isspace())
print("  ".isspace(),end="\n\n")

print("Introducton To Python".istitle())
print("INTRODUCTON TO PYTHON".isupper())
print("hello".startswith("h"))
print("HEllo".swapcase())
print("introducton to python".title())
```

```python
# walrus operator :=
# new to Python 3.8
# assignment expression aka walrus operator
# assigns values to variables as part of a larger expression

happy = True
print(happy)
print("Using Walrus Operator:", happy := True)

foods = []
while True:
    food = input("Enter a food you like (Q to quit): ").upper()
    if food == "Q" :
        break
    foods.append(food)
print(foods)

print()
print("USING WALRUS OPERATOR (:=)")
print()

drinks = []
while (drink := input("Enter a drink you like (Q to quit): ").upper()) != "Q" :
    drinks.append(drink)
print(drinks)
```

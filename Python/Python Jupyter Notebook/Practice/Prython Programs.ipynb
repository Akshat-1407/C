```python
# Program to find the roots of a quadratic equation ax^2 + bx + c = 0.

import math

a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b: "))
c = int(input("Enter the value of c: "))

discriminant = (b*b)-(4*a*c)

if a==0:
    print("Value of a should not be zero. It is not a quadratic equation.")
elif discriminant>0:
    root_1 = round((-b+math.sqrt(discriminant))/(2*a), 2)
    root_2 = round((-b-math.sqrt(discriminant))/(2*a), 2)
    print("Roots are REAL and UNEQUAL.")
    print(f"First root is {root_1}")
    print(f"Second root is {root_2}")
elif discriminant==0:
    root_1 = -b/(2*a)
    root_2 = -b/(2*a)
    print("Roots are REAL and EQUAL.")
    print(f"First root is {root_1}")
    print(f"Second root is {root_2}")
else:
    print("Roots are COMPLEX and IMAGINARY.")
```

```python
# Program to calculate the area of a triangle using Heron's formula.

import math
a=float(input("Enter first side (cm): "))
b=float(input("Enter second side (cm): "))
c=float(input("Enter third side (cm): "))
s=(a+b+c)/2
area=math.sqrt(s*(s-a)*(s-b)*(s-c))
print(f"The area of the triangle is {round(area, 2)} cm")
```

```python
# Program to find compound interest.

principle=0
rate=0
time=0

while True:
    principle=float(input("Enter the principal amount: "))
    if principle<0:
        print("Principle can't be negative.")
    else:
        break

while True:
    rate=float(input("Enter the rate of interest: "))
    if rate<0:
        print("Rate of interest can't be negative.")
    else:
        break

while True:
    time=float(input("Enter the time of interest: "))
    if time<0:
        print("Time can't be negative.")
    else:
        break

amount=principle*pow(1+rate/100, time)
print(f"Your amount is ${round(amount, 2)}. ")
    
```

```python
menu = {
    "Pizza": 4.00, 
    "Burger": 3.00, 
    "Popcorn": 6.00, 
    "Cold drink": 2.00,
    "Soda": 2.00,
    "Lemonade": 1.00
}

cart = []
total = 0

print("------- MENU -------")
for item, price in menu.items():
    print(f"{item:14}: ${price:.2f}")
print("--------------------")

while True:
    food = input("Enter the food from the menu (Q to quit): ").capitalize()
    if food == "Q":
        break
    elif food in menu:
        cart.append(food)
    else:
        print("Sorry, that item is not on the menu. Please try again.")

print("Items in your cart:")
for item in cart:
    print(f"- {item}")

for item in cart:
    total += menu[item]

print(f"Your total is ${total:.2f}.")
```

```python
# number guessing game.

import random
low = 1
high = 100
number = random.randint(low, high)
guesses = 0

while True:
    guess=int(input(f"Enter a number between ({low}-{high}): "))
    guesses += 1
    
    if guess>number:
        print("The guessed number is greater than actual number.")
    elif guess<number:
        print("The guessed number is less than actual number.")
    else:
        print("You guessed correct.")
        break

print(f"This round took you {guesses} guesses.")
```

```python
# Quiz Game

questions = ("How many elements are in the periodic table?: ",
             "Which animal lays the largest eggs?: ",
             "What is the most abundant gas in Earth's atmosphere?: ",
             "How many bones are in the human body?: ",
             "Which planet in the solar system is the hottest?: ")

options = (("A. 116", "B. 117", "C. 118", "D. 119"),
           ("A. Whale", "B. Crocodile", "C. Elephant", "D. Ostrich"),
           ("A. Nitrogen", "B. Oxygen", "C. Carbon-Dioxide", "D. Hydrogen"),
           ("A. 206", "B. 207", "C. 208", "D. 209"),
           ("A. Mercury", "B. Venus", "C. Earth", "D. Mars"))

answers = ("D", "D", "A", "C", "B")
question_no = 0
score = 0


for question in questions:
    print(question)
    for option in options[question_no]:
        print(option)

    guess = input("Enter a guess: ").upper()
    
    if guess == answers[question_no]:
        print("CORRECT ANSWER !")
        score += 1
    else:
        print("INCORRECT ANSWER !")
        print("CORRECT ANSWER : ", answers[question_no])
        
    question_no += 1
    print()

print("Your Score is:", score)
```

```python
# Encryption program in Python

import random
import string

chars = " " + string.punctuation + string.digits + string.ascii_letters
chars = list(chars)
key = chars.copy()

random.shuffle(key)

#ENCRYPT
plain_text = input("Enter a message to encrypt: ")
encrypted_text = ""

for letter in plain_text:
    index = chars.index(letter)
    encrypted_text += key[index]

print(f"Original Message : {plain_text}")
print(f"Encrypted Message: {encrypted_text}")

#DECRYPT
encrypted_text = input("Enter the encrypted message to decrypt: ")
plain_text = ""

for letter in encrypted_text:
    index = key.index(letter)
    plain_text += chars[index]

print(f"Encrypted Message: {encrypted_text}")
print(f"Original Message : {plain_text}")
```

```python
# Return = statement used to end a function and send a result back to the caller.

def add(x, y):
    z = x + y
    return z

def subtract(x, y):
    z = x - y
    return z

def multiply(x, y):
    z = x * y
    return z

def divide(x, y):
    z = x / y
    return z

print(add(2, 1))
print(subtract(2, 1))
print(multiply(2, 1))
print(divide(2, 1))
```

```python
# Return = statement used to end a function and send a result back to the caller

def create_name(first, last):
    first = first.capitalize()
    last = last.capitalize()
    return first + " " + last

full_name = create_name("spongebob", "squarepants")
print(full_name)
```

```python
# default arguments are used after positional arguments.

import time

def count(start, end):
    for i in range(start, end+1):
        print(i)
        time.sleep(1)
    print("DONE !")

count(0, 10)
```

```python
# Keyword arguments = an argument preceded by an identifier
#                     helps with readibility.
#                     order of arguments doesn't matter

def intro(greeting, title, first_name, last_name):
    print(f"{greeting} {title} {first_name} {last_name}")

intro("Hello", last_name="Wick", title="Mr.", first_name="John")

# we can change the position of the arguments but the positional arguments <greeting-Hello> should be at the first.
```

```python
# *args (arguments)               = allows you to pass multiple non key arguments.
# **kwargs (keywaord arguments)   = allows you to pass multiple key word arguments.
#                                   * unpacking operator

# def add(a, b):
#     return a + b
# print(add(1, 2, 3))   < add() takes 2 positional arguments but 3 were given >

def Type(*args):        # after using *args it becomes a tuple and we can itereate over it.
    return(type(args))
print(Type())

def add(*args):
    total = 0
    for arg in args:
        total += arg
    return total
        
print(add(1, 2, 3, 4))

def full_name(*name):
    for any_name in name:
        print(any_name, end=" ")
        
print(full_name("Dr.", "Spongebob", "Squarepants", "III"))
```

```python
# ---- **KWARGS ----

def Type(**kwargs):        # after using **kwargs it becomes a dictianary and we can itereate over it.
    return(type(kwargs))
print(Type())
```

```python
# ---- **KWARGS ----
print_address(street="123 Fake St.",
              pobox="P.O Box 777",
              city="Detroit",
              state="MI",
              zip="54321")

def print_address(**kwargs):
    for value in kwargs.values():
        print(value, end=" ")

# ---- EXERCISE ----
def shipping_label(*args, **kwargs):
    for arg in args:
        print(arg, end=" ")
    print()

    if "apt" in kwargs:
        print(f"{kwargs.get('street')} {kwargs.get('apt')}")
    elif "pobox" in kwargs:
        print(f"{kwargs.get('street')}")
        print(f"{kwargs.get('pobox')}")
    else:
        print(f"{kwargs.get('street')}")

    print(f"{kwargs.get('city')}, {kwargs.get('state')} {kwargs.get('zip')}")

shipping_label("Dr.", "Spongebob", "Squarepants",
               street="123 Fake St.",
               pobox="PO box #1001",
               city="Detroit",
               state="MI",
               zip="54321")
```

```python
# Recursion = We can call a function within the function.
# Docstrings = Tells us about a function or a class. It is similar to comment but not exactly same.

def factorial(n):
    '''Used to find factorial of a number'''
    if n==0 or n==1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))

print("We can call a docstring using __doc__   :-")
print(factorial.__doc__)
```

```python
def fibonacci_sequence(n):
    if n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        fib_seq = fibonacci_sequence(n - 1)
        fib_seq.append(fib_seq[-1] + fib_seq[-2])
        return fib_seq

# Test the function
num_terms = int(input("Enter the number of terms in the Fibonacci sequence: "))
fib_sequence = fibonacci_sequence(num_terms)
print("Fibonacci sequence up to", num_terms, "terms:")
print(fib_sequence)
```

```python
# Higher Order Function =  a function that either:
#                       1. accepts a function as an argument
#                       2. returns a function
#                       (In python, functions are also treated as objects)

print("---- 1. accepts a function as an argument ----") 
def loud(anything):
   return anything.upper()

def quiet(anything):
   return anything.lower()

def hello(func):
   text = func("Hello")
   print(text)

hello(loud)
hello(quiet)

print("----------- 2. returns a function ------------") 
def dividend(y):
    def divisor(x):
        return y/x
    return divisor

num = dividend(10)
print(num(2))
```

```python
# lambda function = function written in 1 line using lambda keyword
#                   accepts any number of arguments, but only has one expression.
#                   (think of it as a shortcut)
#                   (useful if needed for a short period of time, throw-away)

# lambda parameters:expression

def double(x):
    return x*2
print(double(2))

double = lambda x: x * 2
print(double(2))

multiply = lambda x, y: x * y
print(multiply(3,2))

add = lambda x, y, z: x + y + z
print(add(2, 4, 6))

full_name = lambda first_name, last_name: first_name+" "+last_name
print(full_name("Bro","Code"))
```

```python
# List  =  []  ordered and changable. Duplicates ok

# print(dir(fruits))  # helps us to know what we can do with our list.
# print(help(fruits)) # tells us about different functions we can use on our list.

fruits=["apple", "oranges", "banana", "coconut"]
vegetables=["onion", "gourd", "ladyfinger", "cabbage"]

print("apple" in fruits)
print("pineapple" in fruits)

fruits[0] = "pineapple"  # shows list are changable
for fruit in fruits:
    print(fruit, end=" ")
print()

fal = fruits.copy()  # used to create a absolute copy of a list.
print(fal)

fruits.insert(2, "muskmelon")  # used to insert object in the list at a desired point.
print(fruits)

fruits.sort()   # arrange the list in alphabetical order.
print(fruits)

fruits.reverse()  # reverse the elements of the list
print(fruits)

print(fruits.count("oranges")) # used to count the number of specified elements in the list.

print(fruits.index("banana"))  # used to find the index of the specified element in the list.

fruits.append("watermelon")  # used to add another object at the end of the list.
print(fruits)

fruits.remove("watermelon")  # used to remove object in the list.
print(fruits)

fruits.pop(0)   # used to remove an item from the list but is different from remove because it used index as an argument.
print(fruits)

fruits.extend(vegetables)  # used to add another list at the end of the list.
print(fruits)

fruits.clear()
print(fruits)
```

```python
# Set   =  {}  unordered and immutable but add, remove OK, NO duplicates 

vegetables={"onion", "gourd", "ladyfinger", "cabbage"}
print(vegetables)

vegetables.add("garlic") # an element is added randomly at any place.
print(vegetables)

vegetables.remove("garlic") # an element is removed.
print(vegetables)

print(vegetables.pop())  # randomly shows any element in the set

vegetables.clear()
print(vegetables)

# We can't use .index in set because they are unordered. They will not be able to give us the values of the index because elements in a set are placed randomly.
```

```python
# 2-D lists, tuples, sets

fruits     =  ["apple", "oranges", "banana", "coconut"]
vegetables =  ["onion", "ladyfinger", "cabbage"]
pulses     =  ["chana", "rajma", "chole"]

groceries = [fruits, vegetables, pulses]
print(groceries)

print(groceries[0])
print(groceries[1][1])  # first [] represents rows and second [] represents columns.
print()

for collection in groceries:
    for items in collection:
        print(items, end=(" "))
    print()
```

```python
# exercise on 2-D list - numeric pad.

num_pad = ((1, 2, 3),
           (4, 5, 6),
           ("*", 0, "#"))

for rows in num_pad:
    for num in rows:
        print(num, end=(" "))
    print()
```

```python
# Dictionary: It is a collection of {key:value} pairs.
#             ordered and unchangable. NO duplicates.
#             no two keys can be same.

capitals = {"USA" : "Washington DC",
           "India" : "New Delhi",
           "Japan" : "Tokyo",
           "Korea" : "Seoul"}

# print(dir(capitals))  # helps us to know what we can do with our dictionary.
# print(help(capitals)) # tells us about different functions we can use on our dictionary.

print(capitals.get("USA")) # used to get the value of key.

if capitals.get("China"):
    print("The country exists.")
else:
    print("The country dosen't exists.")

print(capitals)
capitals.update({"Korea" : "Hong Kong"})      # used to change the value of the key.
print(capitals)
capitals.update({"Mexoco" : "Mexico City"})   # used to add a new key value pair.
print(capitals)

capitals.pop("Korea")   # used to remove a key value pair.
print(capitals)

capitals.popitem()
print(capitals)         # used to remove a key value pair at the end of the dictionary.
print("\n")

keys=capitals.keys()    # used to pick out all the keys in the dictionary.
print(keys)
for key in keys:
    print(key, end=" ")
print("\n")

values=capitals.values()  # used to pick out all the values in the dictionary.
print(values)
for value in values:
    print(value, end=" ")
print("\n")

items=capitals.items()   # used to pick out all the key value pairs in the dictionary.
print(items)
for key,value in items:
    print(f"{key}:{value}")

capitals.clear()
print(capitals)
```

```python
# sort(key, reverse) method   = used with lists
# sorted(key, reverse) function = used with iterables

students = ["Squidward", "Sandy", "Patrick", "Spongebob", "Mr. Krabs"] 

students.sort()
for i in students:
    print(i)

print()

students.sort(reverse=True)
for i in students:
    print(i)
```

```python
# sort(key, reverse) method   = used with lists
# sorted(key, reverse) function = used with iterables

students = (("Squidward", "F", 60),
            ("Sandy", "A", 33),
            ("Patrick","D", 36),
            ("Spongebob","B", 20),
            ("Mr.Krabs","C", 78))

# Students sorted on the basis of names in alphabetical order.
sorted_students = sorted(students)
for x in sorted_students:
    print(x)

print()

# Students sorted on the basis of names in reverse alphabetical order.
sorted_students = sorted(students, reverse=True)
for x in sorted_students:
    print(x)

print()

# Students sorted on the basis of grade
grade = lambda grades : grades[1]
sorted_students = sorted(students, key=grade)
for x in sorted_students:
    print(x)
print()

# Students sorted on the basis of age
age = lambda ages : ages[2]
sorted_students = sorted(students, key=age)
for x in sorted_students:
    print(x)
```

```python
# map() =   applies a function to each item in an iterable (list, tuple, etc.)
# map(function,iterable)

store = [("shirt",20.00),
         ("pants",25.00),
         ("jacket",50.00),
         ("socks",10.00)]
# using def
def to_rupees(x):
    return (x[0], x[1]*70)
store_rupees =list(map(to_rupees, store))
print(store_rupees)

# using lambda function
store_rupees =list(map(lambda item: (item[0], item[1]*70), store))
print(store_rupees)
```

```python
# filter() =    creates a collection of elements from an iterable, for which a function returns true
# filter(function, iterable)

friends = [("Rachel",20),
         ("Squarepants",25),
         ("Spongebob",17),
         ("Bob",10),
          ("Henry", 18)]
age = lambda data: data[1] >= 18
drinking_friends = list(filter(age, friends))

for name in drinking_friends:
    print(name)
```

```python
# reduce() = apply a function to an iterable and reduce it to a single cumulative value.
#           performs function on first two elements and repeats process until 1 value remains
# reduce(function, iterable)

import functools

letters = ["H","E","L","L","O"]
word = functools.reduce(lambda x, y,:x + y,letters)
print(word)

factorial = [5,4,3,2,1]
result = functools.reduce(lambda x, y,:x * y,factorial)
print(result)
```

```python
# list comprehension =  a way to create a new list with less syntax
# can mimic certain lambda functions, easier to read
# list = [expression for item in iterable]
# list = [expression for item in iterable if conditional]
# list = [expression if/else for item in iterable]

print("------------------------- Without List Comprehension ------------------------")

squares = []                # create an empty list
for i in range(1,11):       # create a for loop
    squares.append(i * i)    # define what each loop iteration should do
print(squares)

print("-------------------------- With List Comprehension --------------------------")

squares = [i * i for i in range(1,11)]     # list = [expression(i*i) for item(for i) in iterable]
print(squares)

print("-----------------------------------------------------------------------------")

marks = [100,54,96,12,86,0,69]     # list = [expression for item in iterable if conditional]
passed_students = [i for i in marks if i>=60]
print(passed_students)

print("-----------------------------------------------------------------------------")

marks = [100,54,96,12,86,0,69]     # list = [expression if/else for item in iterable]
passed_students = [i if i>=60 else "Failed" for i in marks]
print(passed_students)
```

```python
# dictionary comprehension = create dictionaries using an expression
#                            can replace for loops and certain lambda functions
# dictionary = {key: expression for (key,value) in iterable}
# dictionary = {key: expression for (key,value) in iterable if conditional}
# dictionary = {key: (if/else) for (key,value) in iterable}
# dictionary = {key: function(value) for (key,value) in iterable}

temp_in_F = {'New York': 32, 'Boston': 75, 'Los Angeles': 100, 'Chicago': 50}     # dictionary = {key: expression for (key,value) in iterable}
temp_in_C = {key: round((value-32)*(5/9), 2) for (key,value) in temp_in_F.items()}
print(temp_in_C)


weather = {'New York': "snowing", 'Boston': "sunny", 'Los Angeles': "sunny", 'Chicago': "cloudy"}     # dictionary = {key: expression for (key,value) in iterable if conditional}
sunny_weather = {key: value for (key,value) in weather.items() if value == "sunny"}
print(sunny_weather)

cities = {'New York': 32, 'Boston': 75, 'Los Angeles': 100, 'Chicago': 50}     # dictionary = {key: (if/else) for (key,value) in iterable}
description_cities = {key: ("WARM" if value >= 40 else "COLD") for (key,value) in cities.items()}
print(description_cities)

def check_temp(value):
    if value >= 70:
        return "HOT"
    elif 69 >= value >= 40:
        return "WARM"
    else:
        return "COLD"

cities = {'New York': 32, 'Boston': 75, 'Los Angeles': 100, 'Chicago': 50}     # dictionary = {key: function(value) for (key,value) in iterable}
description_cities = {key: check_temp(value) for (key,value) in cities.items()}
print(description_cities)
```

```python
# zip(*iterables) =  aggregate elements from two or more iterables (list, tuples, sets, etc.)
#                    creates a zip object with paired elements stored in tuples for each element

usernames = ["Dude", "Bro", "Mister"]
passwords = ("p@ssword", "abc123", "guest")
login_dates = ["1/1/2021","1/2/2021","1/3/2021"]

# --------------------------------------
users = list(zip(usernames,passwords))

for i in users:
    print(i)
print()
# --------------------------------------
users = dict(zip(usernames,passwords))

for key,value in users.items():
    print(key+" : "+value)
print()
# --------------------------------------
users = zip(usernames,passwords,login_dates)

for i in users:
    print(i)
```
